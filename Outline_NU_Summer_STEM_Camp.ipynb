### Extracting Meta Data from Images with _Exif Tool

The following seciton will install tools to allow the user to extract metadata from images.

```python
# Install exiftool for reading image metadata
!apt-get install -y exiftool -qq


!pip install pyexiftool -qq
```

```python
# Download an image to the local environment using wget
!wget https://raw.githubusercontent.com/thatraghavarora/exif-sample-meta-data/refs/heads/main/gps/DSCN0025.jpg
```

```python
import exiftool

with exiftool.ExifToolHelper() as et:
  metadata = et.get_metadata('DSCN0025.jpg')
  for key, value in metadata[0].items():
    print(key, value)
```
